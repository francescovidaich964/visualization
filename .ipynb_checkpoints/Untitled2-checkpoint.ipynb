```python
from sklearn.datasets import fetch_openml # MNIST data
X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
X_train = X[:6000]
y_train = y[:6000]
```

```python
import numpy as np
y_train = np.array(y_train, dtype='int')
```

```python
from time import time
from sklearn import manifold

tsne = manifold.TSNE(n_components=2, init='random', n_iter = 6000, verbose = 1)

t0 = time()
X_tsne = tsne.fit_transform(X_train)
print(time()-t0)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y_train, cmap='jet')#plt.cm.get_cmap('Reds', 10))
plt.colorbar()

```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10).fit_predict(X_tsne)
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=kmeans, cmap='jet')#plt.cm.get_cmap('Reds', 10))
plt.colorbar()
#c=y_pred
```

```python
from sklearn import mixture
from matplotlib.colors import LogNorm
clf = mixture.GaussianMixture(n_components=10, covariance_type='full')
clf.fit(X_tsne)

x = np.linspace(-120., 130.)
y = np.linspace(-130., 170.)
X, Y = np.meshgrid(x, y)
XX = np.array([X.ravel(), Y.ravel()]).T
Z = -clf.score_samples(XX)
Z = Z.reshape(X.shape)

CS = plt.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0),
                 levels=np.logspace(1, 20, 300))
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y_train, cmap='jet')#plt.cm.get_cmap('Reds', 10))
plt.colorbar()
```

```python
X_train = X[:6000]
y_train = y[:6000]
y_train = np.array(y_train, dtype='int')
```

```python
from scipy import linalg as la
import numpy as np
#for i in range(X_train.shape[0]):
#    X_train[i] = (X_train[i] - X_train[i].mean())/X_train[i].std()
    
# PCA 
C = np.cov(X_train.T)
U, spectrum, Vt = la.svd(C)
print ('Spectrum: ', spectrum[:100].sum()/spectrum.sum(),'\n')
#print(spectrum[500:550])
#perc_expl2 = (spectrum[0] + spectrum[1]) / spectrum.sum()
#print('Percent of the total variability explained considering the two main features: ', perc_expl2, '\n')
```

```python
Xp = np.dot(Vt,X.T)
Xp[150:,:] = 0
Xpp = np.dot(U, Xp)
```

```python
X_train = Xp[:150]
X_train.T[:1000].shape
```

```python
from time import time
from sklearn import manifold
perplexities = [2, 10, 30, 50, 100, 150]
X_tsnes = []
for perplexity in perplexities:
    tsne = manifold.TSNE(n_components=2, init='random', perplexity = perplexity, n_iter = 5000, learning_rate=100, verbose = 0)
    #t0 = time()
    X_tsnes.append(tsne.fit_transform(X_train.T[:1000]))
    #print(time()-t0)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
for i in range(len(X_tsnes)):
    plt.scatter(X_tsnes[i][:,0], X_tsnes[i][:,1], c=y_train[:1000], cmap='jet')#plt.cm.get_cmap('Reds', 10))
    plt.colorbar()
    plt.title("Perplexity: " + str(perplexities[i]))
    plt.show()
    plt.close()
```
