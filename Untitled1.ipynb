```r
x  <- c(15.8, 15.9, 16, 16.1, 16.2)
p1 <- c(0.15, 0.21, 0.35, 0.15, 0.14)
p2 <- c(0.14, 0.05, 0.64, 0.08, 0.09)
```

```r
E1 <- sum(x * p1)
E2 <- sum(x * p2)
Var1 <- sum(x**2 * p1) - E1**2
Var2 <- sum(x**2 * p2) - E2**2
```

```r
print(E1)
print(E2)
print(Var1)
print(Var2)
```

```r
x <- seq(-0.1, 150, 0.1)
yexp <- dexp(x, rate = 1/30)

par(mfrow=c(2,1), mgp=c(2.0, 0.8, 0), mar=c(3.5,3.5, 1,1))
plot(x,x, type='n', xlim=c(0,150), ylim=c(0,0.04), xlab="behbeh",ylab="P(x | p,n)")
#points(x, b1, pch=20, col="red")
lines(x, yexp, type='s', lty=1, col="red", lwd=1)
```
